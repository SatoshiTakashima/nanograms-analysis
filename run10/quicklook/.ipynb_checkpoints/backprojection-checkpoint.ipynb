```python
import numpy as np
import matplotlib.pyplot as plt


def normalize_vectors(v):
    v = np.asarray(v, dtype=float)
    norm = np.linalg.norm(v, axis=1, keepdims=True)
    if np.any(norm == 0):
        raise ValueError("zero-length vector is included")
    return v / norm


def cone_axis_orthogonal_basis(cone_axis):
    cone_axis = normalize_vectors(cone_axis)

    ref = np.tile(np.array([0.0, 0.0, 1.0]), (len(cone_axis), 1))
    nearly_z = np.abs(cone_axis[:, 2]) > 0.9
    ref[nearly_z] = np.array([1.0, 0.0, 0.0])

    u1 = np.cross(cone_axis, ref)
    u1 = normalize_vectors(u1)

    u2 = np.cross(cone_axis, u1)
    u2 = normalize_vectors(u2)

    return u1, u2


def plane_basis_from_normal(plane_normal):
    n = np.asarray(plane_normal, dtype=float)
    n = n / np.linalg.norm(n)

    if abs(n[0]) > 0.9:
        axis_u = np.array([0.0, 1.0, 0.0])
    else:
        axis_u = np.cross(n, np.array([1.0, 0.0, 0.0]))
        axis_u = axis_u / np.linalg.norm(axis_u)

    axis_v = np.cross(n, axis_u)
    axis_v = axis_v / np.linalg.norm(axis_v)

    return axis_u, axis_v


def project_compton_cones_to_plane(
    cone_vertex,
    cone_axis,
    theta_array,
    plane_point,
    plane_normal,
    num_phi=720,
    forward_only=True,
):
    cone_vertex = np.asarray(cone_vertex, dtype=float)
    cone_axis = normalize_vectors(cone_axis)
    theta_array = np.asarray(theta_array, dtype=float)
    plane_point = np.asarray(plane_point, dtype=float)
    plane_normal = np.asarray(plane_normal, dtype=float)
    plane_normal = plane_normal / np.linalg.norm(plane_normal)

    u1, u2 = cone_axis_orthogonal_basis(cone_axis)

    phi = np.linspace(0.0, 2.0 * np.pi, num_phi, endpoint=False)
    cos_phi = np.cos(phi)[None, :, None]
    sin_phi = np.sin(phi)[None, :, None]

    cos_theta = np.cos(theta_array)[:, None, None]
    sin_theta = np.sin(theta_array)[:, None, None]

    cone_dir = (
        cos_theta * cone_axis[:, None, :]
        + sin_theta * (cos_phi * u1[:, None, :] + sin_phi * u2[:, None, :])
    )

    numerator = np.sum(
        (plane_point[None, :] - cone_vertex) * plane_normal[None, :],
        axis=1,
    )[:, None]

    denominator = np.sum(cone_dir * plane_normal[None, None, :], axis=2)

    valid = np.abs(denominator) > 1.0e-12
    t = np.where(valid, numerator / denominator, np.nan)

    if forward_only:
        valid = valid & (t > 0.0)

    points_3d = cone_vertex[:, None, :] + t[:, :, None] * cone_dir

    plane_axis_u, plane_axis_v = plane_basis_from_normal(plane_normal)
    rel = points_3d - plane_point[None, None, :]

    plane_u = np.sum(rel * plane_axis_u[None, None, :], axis=2)
    plane_v = np.sum(rel * plane_axis_v[None, None, :], axis=2)

    return plane_u, plane_v, points_3d, valid

# 使用例
pos_hit1 = np.stack([dataDict[f"hit1_pos{xyz}"] for xyz in ["x", "y", "z"]], axis=1)
pos_hit2 = np.stack([dataDict[f"hit2_pos{xyz}"] for xyz in ["x", "y", "z"]], axis=1)

L21 = pos_hit1 - pos_hit2
cone_axis = normalize_vectors(L21)
theta_array = dataDict["theta_k"]

plane_point = np.array([50.0, 0.0, 0.0])
plane_normal = np.array([1.0, 0.0, 0.0])

plane_u, plane_v, points_3d, valid = project_compton_cones_to_plane(
    cone_vertex=pos_hit1,
    cone_axis=cone_axis,
    theta_array=theta_array,
    plane_point=plane_point,
    plane_normal=plane_normal,
    num_phi=1000,
)

u = plane_u[valid]
v = plane_v[valid]

plt.figure(figsize=(6, 5))
plt.hist2d(u, v, bins=100, range=[[-100, 100], [-100, 100]])
plt.xlabel("plane u (cm)")
plt.ylabel("plane v (cm)")
plt.gca().set_aspect("equal")
plt.colorbar(label="counts")
plt.show()
```
