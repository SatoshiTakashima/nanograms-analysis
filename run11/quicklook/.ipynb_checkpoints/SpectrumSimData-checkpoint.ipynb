```python
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import uproot
```

```python
numFEC = 4
```

```python
dirFile = Path("/Users/takashima/work/grams/nanograms-analysis/run11/simulation-scan/sim_results_param1_0p020-1e9")
fileName = "simulation_Na22_merged.root"
tFile = uproot.open(dirFile/fileName)

branchList = ["energy"]
energyArrayList = [tFile["hittree"].arrays(branchList, f"(num_hits==1)&(detector=={i})", library="np") for i in range(numFEC)]
```

```python
binsArray = np.linspace(200, 1400, 61)
countsList = np.array([np.histogram(energyArrayList[i]["energy"], bins=binsArray)[0] for i in range(numFEC)])
countAllList = np.sum(countsList, axis=0)
```

```python
fig, ax = plt.subplots(1,1,figsize=(6,5))
for i in range(numFEC):
    ax.stairs(countsList[i], binsArray, label=f"FEC{i}")
ax.stairs(countAllList, binsArray, label="All", color="k")
ax.legend()
ax.set_xlabel("Energy (keV)")
ax.set_ylabel("Count")
ax.grid()
```
